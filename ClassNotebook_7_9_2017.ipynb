```python
import re
import csv
import networkx as nx
import matplotlib.pyplot as plt

t = open('/home/faculty/millbr02/pub/un-debates-augmented-1.csv','r',encoding='latin-1')
csv_iter = csv.reader(t)
list_1 = []
db = []
for row in csv_iter:
        db.append(row)
        list_1.append(row[4])

import psycopg2
conn = psycopg2.connect("postgresql://millbr02:@localhost/world")
curs = conn.cursor()
curs.execute("select * from country limit 400")
res = curs.fetchall()
dict1 = {}
for row in res:
    list_1.append(row[1])


neo = {x: {} for x in list_1}
for k in neo:
    neo[k] = {x:0 for x in list_1}
    



for row in db[1:]:
    text = row[3]
    country_speaking = row[4]
    for word in list_1:
        count = text.count(word)
        neo[country_speaking][word] = count



def draw_graph(graph, labels=None, graph_layout='shell',
               node_size=1600, node_color='blue', node_alpha=0.3,
               node_text_size=12,
               edge_color='blue', edge_alpha=0.3, edge_tickness=1,
               edge_text_pos=0.3,
               text_font='sans-serif'):

    # create networkx graph
    G=nx.Graph()

    # add edges
    for edge in graph:
        G.add_edge(edge[0], edge[1])

    # these are different layouts for the network you may try
    # shell seems to work best
    if graph_layout == 'spring':
        graph_pos=nx.spring_layout(G)
    elif graph_layout == 'spectral':
        graph_pos=nx.spectral_layout(G)
    elif graph_layout == 'random':
        graph_pos=nx.random_layout(G)
    else:
        graph_pos=nx.shell_layout(G)

    # draw graph
    nx.draw_networkx_nodes(G,graph_pos,node_size=node_size, 
                           alpha=node_alpha, node_color=node_color)
    nx.draw_networkx_edges(G,graph_pos,width=edge_tickness,
                           alpha=edge_alpha,edge_color=edge_color)
    nx.draw_networkx_labels(G, graph_pos,font_size=node_text_size,
                            font_family=text_font)

    if labels is None:
        labels = range(len(graph))

    edge_labels = dict(zip(graph, labels))
    nx.draw_networkx_edge_labels(G, graph_pos, edge_labels=edge_labels, 
                                 label_pos=edge_text_pos)

    # show graph
    plt.show()


```
